```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import Markdown as md

# --- Select event window and countries ---
event_start, event_end = 1993, 2000
countries = ["United States", "Russian Federation", "Poland", "United Kingdom", "Germany"]

subset = df[(df["country"].isin(countries)) &
            (df["year"].between(event_start, event_end))].copy()

# ---------------- Plot : % of GDP ----------------
plt.figure(figsize=(12,6))
sns.lineplot(data=subset, x="year", y="mil_gdp_share", hue="country", marker="o")
for yr in [1994, 1998, 1999, 2000]:
    plt.axvline(x=yr, color="gray", linestyle="--", alpha=0.6)

plt.title("1993â€“2000: Postâ€“Cold War Adjustment & NATO Expansion\nDefense Spending (% of GDP)")
plt.ylabel("Military Expenditure (% of GDP)")
plt.xlabel("Year")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(title="Country", loc="upper right")
plt.tight_layout()
plt.show()

# ---------------- Explanation (simple + intuitive) ----------------
explanation = """
**Understanding the Trends (1993â€“2000):**

- **1993â€“1994:** Russiaâ€™s spending is still high but unstable; the economy is chaotic after the Soviet collapse.  
  Western countries, especially the U.S. and Europe, are cutting back because they feel secure in the new postâ€“Cold War world.

- **1995â€“1997:** Everyoneâ€™s defense budgets keep shrinking. The West enjoys a â€œpeace dividend,â€ 
  while Russia struggles with weak finances and unpaid soldiers.

- **1998:** Russia hits a deep economic crisis; the ruble crashes and defense spending falls sharply.  
  This drop is visible in both GDP % and USD terms.

- **1999:** NATO adds Poland, Hungary, and the Czech Republic. The West does *not* raise its budgets yet. 
  Expansion was political, not military. Russia begins a small rebound as it stabilizes internally.

- **2000:** Putin comes to power. Russiaâ€™s spending curve starts to rise again, marking the start of its military recovery phase.
"""
display(md(explanation))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import Markdown as md

# --- Select window ---
event_start, event_end = 2017, 2021
subset = df[df["year"].between(event_start, event_end)].copy()

# --- Grouping ---
country_groups = {
    "United States": ["United States"],
    "China": ["China"],
    "Russia": ["Russian Federation"],
    "Western Allies": ["United Kingdom", "France", "Germany", "Australia"],
    "Asian Powers": ["Japan", "Korea, Rep.", "India"],
    "Middle East Bloc": ["Saudi Arabia", "Iran, Islamic Rep.", "Turkiye"]
}

# Build group averages (log scale for comparability)
grouped_data = []
for group, members in country_groups.items():
    temp = subset[subset["country"].isin(members)].groupby("year")["mil_usd"].mean().reset_index()
    temp["country_group"] = group
    temp["log_usd"] = np.log1p(temp["mil_usd"])
    grouped_data.append(temp)

grouped = pd.concat(grouped_data, ignore_index=True)

# --- Plot ---
plt.figure(figsize=(12,6))
sns.lineplot(data=grouped, x="year", y="log_usd", hue="country_group", marker="o")
for yr in [2017, 2018, 2020]:
    plt.axvline(x=yr, color="gray", linestyle="--", alpha=0.5)
plt.title("2017â€“2021: China and Russia as Peer Competitors\n(Log of Military Expenditure, USD)")
plt.ylabel("Log of Military Expenditure (USD)")
plt.xlabel("Year")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend(title="Country / Group", loc="upper left")
plt.tight_layout()
plt.show()

# --- Explanations ---
explanation = """
### Understanding the Trends (2017â€“2021): Country and Group Narratives

#### **United States ðŸ‡ºðŸ‡¸**
- **2017:** Trump administrationâ€™s *National Defense Strategy (2018)* names China and Russia as â€œrevisionist powers.â€
- **2018â€“2019:** Budgets rise sharply under â€œgreat power competitionâ€ doctrine.
- **2020â€“2021:** COVID-19 slows overall growth but spending remains historically high.

#### **China ðŸ‡¨ðŸ‡³**
- Continuous steady climb across all years.
- Driven by military modernization, blue-water naval ambitions, and Taiwan-focused deterrence.
- **2021:** Spending reaches a new highâ€”reflecting strategic confidence and Indo-Pacific ambitions.

#### **Russia ðŸ‡·ðŸ‡º**
- Budget oscillates mildly but remains elevated post-Crimea.
- Economic sanctions cap expansion, but rearmament of missile and air-defense systems continues.
- **2020â€“2021:** Rising tension in Ukraine border regions triggers renewed uptick.

#### **Western Allies (ðŸ‡¬ðŸ‡§ðŸ‡«ðŸ‡·ðŸ‡©ðŸ‡ªðŸ‡¦ðŸ‡º)**
- After years of cuts, defense budgets climb again as NATO adopts a deterrence posture.
- **2019â€“2021:** Stronger U.S. coordination and new commitments to the Indo-Pacific (e.g., AUKUS).

#### **Asian Powers (ðŸ‡¯ðŸ‡µðŸ‡°ðŸ‡·ðŸ‡®ðŸ‡³)**
- **Japan:** Steady incremental growth amid East China Sea tensions.
- **South Korea:** Increases spending to strengthen missile defense.
- **India:** Border clashes with China (2020 Galwan Valley) cause visible rise.

#### **Middle East Bloc (ðŸ‡¸ðŸ‡¦ðŸ‡®ðŸ‡·ðŸ‡¹ðŸ‡·)**
- **Saudi Arabia:** Spending stays high despite oil shocks; Yemen war persists.
- **Iran:** Defense constrained by sanctions, but maintains regional proxies.
- **Turkey:** Spending rises after 2016 coup attempt; emphasizes indigenous defense production.

---
**Overall Insight:**  
This period marks a synchronized rearmament wave â€”  
the U.S. pivots to counter China, Russia rebuilds regional capability,  
and middle powers adapt to an unstable multipolar world.
"""
display(md(explanation))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import Markdown as md

# --- Select full range ---
subset = df[df["year"].between(1993, 2023)].copy()

# --- Group definitions ---
country_groups = {
    "United States": ["United States"],
    "China": ["China"],
    "Russia": ["Russian Federation"],
    "Western Allies": ["United Kingdom", "France", "Germany", "Australia"],
    "Asian Powers": ["Japan", "Korea, Rep.", "India"],
    "Middle East Bloc": ["Saudi Arabia", "Iran, Islamic Rep.", "Turkiye"]
}

# --- Grouping and averaging ---
grouped_data = []
for group, members in country_groups.items():
    temp = subset[subset["country"].isin(members)].groupby("year")["mil_gdp_share"].mean().reset_index()
    temp["country_group"] = group
    grouped_data.append(temp)

grouped = pd.concat(grouped_data, ignore_index=True)

# --- Plot setup ---
plt.figure(figsize=(14,7))
sns.lineplot(data=grouped, x="year", y="mil_gdp_share", hue="country_group", linewidth=2.2)

# Mark major historical phases
phase_breaks = {
    2001: "9/11 & War on Terror",
    2014: "Crimea Annexation",
    2018: "China as Peer Competitor",
    2022: "Ukraine War"
}
for year, label in phase_breaks.items():
    plt.axvline(x=year, color="gray", linestyle="--", alpha=0.5)
    plt.text(year+0.1, 0.3, label, rotation=90, verticalalignment="bottom", fontsize=9, alpha=0.8)

# Title & labels
plt.title("Defense Spending as % of GDP (1993â€“2023)\nEvolution of Global Military Commitment", fontsize=14)
plt.ylabel("Military Expenditure (% of GDP)")
plt.xlabel("Year")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend(title="Country / Group", loc="upper right")
plt.tight_layout()
plt.show()

# --- Explanatory Summary ---
explanation = """
### Multi-Period Interpretation (1993â€“2023)

#### **1993â€“2000 â€” Post-Cold War Retrenchment**
- U.S., Western Europe, and Russia all cut defense budgets.
- â€œPeace dividendâ€ after Soviet collapse; global military expenditure as % of GDP bottoms out.
- **China** begins quiet, steady modernization from a low base.

#### **2001â€“2011 â€” War on Terror Era**
- **U.S.** and allies surge spending after 9/11 â€” Iraq (2003), Afghanistan (2001â€“2011).
- **Middle East Bloc** also spikes â€” Saudi Arabia and Iran maintain regional defense races.
- **Russia** rebuilds under rising oil prices and military reform (late 2000s).

#### **2012â€“2016 â€” Retrenchment & Fiscal Sequestration**
- **U.S.** cuts back (budget sequestration).  
- **Europe** flatlines amid austerity.  
- **China** keeps gradual, disciplined growth.

#### **2014â€“2021 â€” Crimea, Chinaâ€™s Rise, and Strategic Rivalry**
- **Russia:** sharp uptick after annexing Crimea (2014).  
- **NATO & Western Allies:** slow but steady rearmament.  
- **Asia:** Japan, India, and South Korea expand budgets amid regional flashpoints.  
- **China:** rises consistently, symbolizing its emergence as peer competitor.

#### **2022â€“2023 â€” Ukraine War & Indo-Pacific Tensions**
- **Russia:** explodes upward; war economy dominates GDP.  
- **Western Europe:** unprecedented defense increases (Germanyâ€™s Zeitenwende).  
- **U.S. and Asia:** strengthen deterrence.  
- **Global average spending** climbs â€” reversing 30 years of gradual disarmament.

---

**Overall Message:**  
The long arc (1993â€“2023) traces a complete global cycle â€”  
from **disarmament and peace dividend**, to **crisis-driven rearmament**, to **structural militarization**.  

- The **U.S. and its allies** lead major rearmament waves tied to geopolitical shocks.  
- **Russia** oscillates with regime strength and oil prices, culminating in wartime mobilization.  
- **China**â€™s smooth upward curve marks its transition from observer to great power.  
- **Middle Eastern powers** remain consistently high â€” conflict and energy rents anchor spending.  

This historical foundation sets the stage for forecasting:  
**How strongly does rival spending (especially from China or Russia) predict U.S. defense growth?**
"""
display(md(explanation))
```
