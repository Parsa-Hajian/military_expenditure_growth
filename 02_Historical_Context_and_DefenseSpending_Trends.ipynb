```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import Markdown as md

# --- Select event window and countries ---
event_start, event_end = 1993, 2000
countries = ["United States", "Russian Federation", "Poland", "United Kingdom", "Germany"]

subset = df[(df["country"].isin(countries)) &
            (df["year"].between(event_start, event_end))].copy()

# ---------------- Plot : % of GDP ----------------
plt.figure(figsize=(12,6))
sns.lineplot(data=subset, x="year", y="mil_gdp_share", hue="country", marker="o")
for yr in [1994, 1998, 1999, 2000]:
    plt.axvline(x=yr, color="gray", linestyle="--", alpha=0.6)

plt.title("1993–2000: Post–Cold War Adjustment & NATO Expansion\nDefense Spending (% of GDP)")
plt.ylabel("Military Expenditure (% of GDP)")
plt.xlabel("Year")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(title="Country", loc="upper right")
plt.tight_layout()
plt.show()

# ---------------- Explanation (simple + intuitive) ----------------
explanation = """
**Understanding the Trends (1993–2000):**

- **1993–1994:** Russia’s spending is still high but unstable; the economy is chaotic after the Soviet collapse.  
  Western countries, especially the U.S. and Europe, are cutting back because they feel secure in the new post–Cold War world.

- **1995–1997:** Everyone’s defense budgets keep shrinking. The West enjoys a “peace dividend,” 
  while Russia struggles with weak finances and unpaid soldiers.

- **1998:** Russia hits a deep economic crisis; the ruble crashes and defense spending falls sharply.  
  This drop is visible in both GDP % and USD terms.

- **1999:** NATO adds Poland, Hungary, and the Czech Republic. The West does *not* raise its budgets yet. 
  Expansion was political, not military. Russia begins a small rebound as it stabilizes internally.

- **2000:** Putin comes to power. Russia’s spending curve starts to rise again, marking the start of its military recovery phase.
"""
display(md(explanation))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import Markdown as md

# --- Select event window and countries ---
event_start, event_end = 2012, 2016
countries = ["United States", "United Kingdom", "France", "Germany",
             "Australia", "Russian Federation", "Saudi Arabia", "Iran, Islamic Rep."]

subset = df[(df["country"].isin(countries)) &
            (df["year"].between(event_start, event_end))].copy()

key_years = [2012, 2014, 2016]

# ---------------- Plot : Log of USD ----------------
plt.figure(figsize=(12,6))
sns.lineplot(data=subset, x="year", y=np.log1p(subset["mil_usd"]), hue="country", marker="o")
for yr in key_years:
    plt.axvline(x=yr, color="gray", linestyle="--", alpha=0.6)
plt.title("2012–2016: Sequestration and Retrenchment\nDefense Spending (Log of Current USD)")
plt.ylabel("Log of Military Expenditure (USD)")
plt.xlabel("Year")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(title="Country", loc="upper right")
plt.tight_layout()
plt.show()

# ---------------- Country-by-country explanation ----------------
explanation = """
### Understanding the Trends (2012–2016): Country by Country

#### **United States 🇺🇸**
- **2012:** The Budget Control Act (2011) enforces automatic spending cuts—known as *sequestration*. 
  Military spending drops for the first time in a decade.
- **2014:** Budgets bottom out as U.S. forces withdraw from Iraq and Afghanistan. 
  The focus shifts toward fiscal discipline rather than expansion.
- **2016:** Slight recovery as global threats reemerge, but spending remains well below 2010 levels.

#### **United Kingdom 🇬🇧**
- Defense spending shrinks alongside austerity measures.
- **2014:** Russian annexation of Crimea prompts strategic review, halting further decline.
- **2016:** London begins modest rearmament under NATO’s renewed eastern defense commitments.

#### **France 🇫🇷**
- Consistent moderate spending with spikes around **2015**, reflecting interventions in 
  [Mali (Operation Serval)](https://en.wikipedia.org/wiki/Operation_Serval) and anti-terror campaigns after the Paris attacks.
- Domestic terror threats sustain defense budgets despite EU austerity.

#### **Germany 🇩🇪**
- Maintains minimal spending levels, adhering to its fiscal restraint culture.
- **2014:** Crimea crisis sparks domestic debate on rearmament, but budget increases are still limited.

#### **Australia 🇦🇺**
- Ends Afghanistan involvement; temporary decline.
- **2016:** Announces a new [Defence White Paper](https://www.defence.gov.au/about/strategic-policy/defence-white-paper-2016), signaling a pivot toward Indo-Pacific security and gradual budget recovery.

#### **Russia 🇷🇺**
- **2012–2014:** Rapid spending buildup following the [Georgia War (2008)](https://en.wikipedia.org/wiki/Russo-Georgian_War) and modernizing drive.
- **2014:** Annexation of Crimea; defense budget surges despite sanctions.
- **2016:** Military modernization peaks; spending reaches post-Soviet highs.

#### **Saudi Arabia 🇸🇦**
- **2014–2016:** Oil wealth fuels defense surge; enters the Yemen conflict in 2015.
  Spending becomes one of the world’s highest relative to GDP.

#### **Iran 🇮🇷**
- Sustains modest but rising defense budgets despite sanctions.
- **2015:** [Nuclear deal (Check out my paper on JCPOA)](https://www.linkedin.com/pulse/trigger-strategy-international-economics-analysis-parsa-hajiannezhad-a7hse/?trackingId=QbrNc06GTs%2BA3q%2BdPGSu4g%3D%3D) offers economic relief, allowing limited expansion.
- By **2016**, growth stabilizes as oil exports recover slightly.
"""
display(md(explanation))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import Markdown as md

# --- Select window ---
event_start, event_end = 2017, 2021
subset = df[df["year"].between(event_start, event_end)].copy()

# --- Grouping ---
country_groups = {
    "United States": ["United States"],
    "China": ["China"],
    "Russia": ["Russian Federation"],
    "Western Allies": ["United Kingdom", "France", "Germany", "Australia"],
    "Asian Powers": ["Japan", "Korea, Rep.", "India"],
    "Middle East Bloc": ["Saudi Arabia", "Iran, Islamic Rep.", "Turkiye"]
}

# Build group averages (log scale for comparability)
grouped_data = []
for group, members in country_groups.items():
    temp = subset[subset["country"].isin(members)].groupby("year")["mil_usd"].mean().reset_index()
    temp["country_group"] = group
    temp["log_usd"] = np.log1p(temp["mil_usd"])
    grouped_data.append(temp)

grouped = pd.concat(grouped_data, ignore_index=True)

# --- Plot ---
plt.figure(figsize=(12,6))
sns.lineplot(data=grouped, x="year", y="log_usd", hue="country_group", marker="o")
for yr in [2017, 2018, 2020]:
    plt.axvline(x=yr, color="gray", linestyle="--", alpha=0.5)
plt.title("2017–2021: China and Russia as Peer Competitors\n(Log of Military Expenditure, USD)")
plt.ylabel("Log of Military Expenditure (USD)")
plt.xlabel("Year")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend(title="Country / Group", loc="upper left")
plt.tight_layout()
plt.show()

# --- Explanations ---
explanation = """
### Understanding the Trends (2017–2021): Country and Group Narratives

#### **United States 🇺🇸**
- **2017:** Trump administration’s *National Defense Strategy (2018)* names China and Russia as “revisionist powers.”
- **2018–2019:** Budgets rise sharply under “great power competition” doctrine.
- **2020–2021:** COVID-19 slows overall growth but spending remains historically high.

#### **China 🇨🇳**
- Continuous steady climb across all years.
- Driven by military modernization, blue-water naval ambitions, and Taiwan-focused deterrence.
- **2021:** Spending reaches a new high—reflecting strategic confidence and Indo-Pacific ambitions.

#### **Russia 🇷🇺**
- Budget oscillates mildly but remains elevated post-Crimea.
- Economic sanctions cap expansion, but rearmament of missile and air-defense systems continues.
- **2020–2021:** Rising tension in Ukraine border regions triggers renewed uptick.

#### **Western Allies (🇬🇧🇫🇷🇩🇪🇦🇺)**
- After years of cuts, defense budgets climb again as NATO adopts a deterrence posture.
- **2019–2021:** Stronger U.S. coordination and new commitments to the Indo-Pacific (e.g., AUKUS).

#### **Asian Powers (🇯🇵🇰🇷🇮🇳)**
- **Japan:** Steady incremental growth amid East China Sea tensions.
- **South Korea:** Increases spending to strengthen missile defense.
- **India:** Border clashes with China (2020 Galwan Valley) cause visible rise.

#### **Middle East Bloc (🇸🇦🇮🇷🇹🇷)**
- **Saudi Arabia:** Spending stays high despite oil shocks; Yemen war persists.
- **Iran:** Defense constrained by sanctions, but maintains regional proxies.
- **Turkey:** Spending rises after 2016 coup attempt; emphasizes indigenous defense production.

---
**Overall Insight:**  
This period marks a synchronized rearmament wave —  
the U.S. pivots to counter China, Russia rebuilds regional capability,  
and middle powers adapt to an unstable multipolar world.
"""
display(md(explanation))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import Markdown as md

# --- Select the time window ---
event_start, event_end = 2022, 2023
subset = df[df["year"].between(event_start, event_end)].copy()

# --- Define country groups ---
country_groups = {
    "United States": ["United States"],
    "China": ["China"],
    "Russia": ["Russian Federation"],
    "Western Allies": ["United Kingdom", "France", "Germany", "Australia"],
    "Asian Powers": ["Japan", "Korea, Rep.", "India"],
    "Middle East Bloc": ["Saudi Arabia", "Iran, Islamic Rep.", "Turkiye"]
}

# --- Build grouped dataset using % of GDP ---
grouped_data = []
for group, members in country_groups.items():
    temp = subset[subset["country"].isin(members)].groupby("year")["mil_gdp_share"].mean().reset_index()
    temp["country_group"] = group
    grouped_data.append(temp)

grouped = pd.concat(grouped_data, ignore_index=True)

# --- Plot ---
plt.figure(figsize=(12,6))
sns.lineplot(data=grouped, x="year", y="mil_gdp_share", hue="country_group", marker="o")
plt.axvline(x=2022, color="gray", linestyle="--", alpha=0.5)
plt.title("2022–2023: Ukraine War and Indo-Pacific Escalation\nDefense Spending (% of GDP)")
plt.ylabel("Military Expenditure (% of GDP)")
plt.xlabel("Year")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend(title="Country / Group", loc="upper left")
plt.tight_layout()
plt.show()

# --- Explanations ---
explanation = """
### Understanding the Trends (2022–2023): Country and Group Narratives (GDP Share)

#### **United States 🇺🇸**
- Defense spending rises modestly **as a share of GDP**, since the economy also expands.
- Aid to Ukraine and Indo-Pacific deployments push defense above 3% of GDP — up from pre-2020 levels.

#### **Russia 🇷🇺**
- The **sharpest increase** across all nations.
- War mobilization drives military spending to nearly **5–6% of GDP**, levels unseen since the 1990s.
- Reflects a war economy: rising costs amid economic contraction.

#### **China 🇨🇳**
- Despite large absolute budgets, GDP growth keeps the share stable around **1.6–1.7%**.
- This shows China’s methodical, long-term modernization rather than panic spending.

#### **Western Allies (🇬🇧🇫🇷🇩🇪🇦🇺)**
- Clear post-Ukraine rise:  
  - **Germany’s “Zeitenwende”** marks a historic break from its post–Cold War restraint.  
  - **UK and France** move closer to NATO’s **2% of GDP target**.
- **Australia** increases defense-to-GDP ratio for AUKUS investments.

#### **Asian Powers (🇯🇵🇰🇷🇮🇳)**
- **Japan:** Crosses 1% of GDP — a symbolic threshold since WWII.  
- **South Korea:** Near 2.7%, balancing North Korean missile threats.  
- **India:** Slight uptick tied to China border tensions and “Atmanirbhar Bharat” defense drive.

#### **Middle East Bloc (🇸🇦🇮🇷🇹🇷)**
- **Saudi Arabia:** Remains among world leaders, near **7% of GDP** — sustained war in Yemen and regional competition.  
- **Iran:** Modest increase despite sanctions.  
- **Turkey:** High but steady, reflecting operational commitments in Syria and growing drone industry.

---

**Overall Insight:**  
When seen as a share of GDP, 2022–2023 reveals **who feels most threatened** rather than who spends most in dollars.  
- **Russia and Saudi Arabia** show extreme mobilization.  
- **Europe and the U.S.** show determined, but measured rearmament.  
- **China and Asia** stay steady — pacing themselves for a long-term strategic race.  
- **This marks the onset of an entrenched, multi-front global militarization cycle.**
"""
display(md(explanation))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import Markdown as md

# --- Select full range ---
subset = df[df["year"].between(1993, 2023)].copy()

# --- Group definitions ---
country_groups = {
    "United States": ["United States"],
    "China": ["China"],
    "Russia": ["Russian Federation"],
    "Western Allies": ["United Kingdom", "France", "Germany", "Australia"],
    "Asian Powers": ["Japan", "Korea, Rep.", "India"],
    "Middle East Bloc": ["Saudi Arabia", "Iran, Islamic Rep.", "Turkiye"]
}

# --- Grouping and averaging ---
grouped_data = []
for group, members in country_groups.items():
    temp = subset[subset["country"].isin(members)].groupby("year")["mil_gdp_share"].mean().reset_index()
    temp["country_group"] = group
    grouped_data.append(temp)

grouped = pd.concat(grouped_data, ignore_index=True)

# --- Plot setup ---
plt.figure(figsize=(14,7))
sns.lineplot(data=grouped, x="year", y="mil_gdp_share", hue="country_group", linewidth=2.2)

# Mark major historical phases
phase_breaks = {
    2001: "9/11 & War on Terror",
    2014: "Crimea Annexation",
    2018: "China as Peer Competitor",
    2022: "Ukraine War"
}
for year, label in phase_breaks.items():
    plt.axvline(x=year, color="gray", linestyle="--", alpha=0.5)
    plt.text(year+0.1, 0.3, label, rotation=90, verticalalignment="bottom", fontsize=9, alpha=0.8)

# Title & labels
plt.title("Defense Spending as % of GDP (1993–2023)\nEvolution of Global Military Commitment", fontsize=14)
plt.ylabel("Military Expenditure (% of GDP)")
plt.xlabel("Year")
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend(title="Country / Group", loc="upper right")
plt.tight_layout()
plt.show()

# --- Explanatory Summary ---
explanation = """
### Multi-Period Interpretation (1993–2023)

#### **1993–2000 — Post-Cold War Retrenchment**
- U.S., Western Europe, and Russia all cut defense budgets.
- “Peace dividend” after Soviet collapse; global military expenditure as % of GDP bottoms out.
- **China** begins quiet, steady modernization from a low base.

#### **2001–2011 — War on Terror Era**
- **U.S.** and allies surge spending after 9/11 — Iraq (2003), Afghanistan (2001–2011).
- **Middle East Bloc** also spikes — Saudi Arabia and Iran maintain regional defense races.
- **Russia** rebuilds under rising oil prices and military reform (late 2000s).

#### **2012–2016 — Retrenchment & Fiscal Sequestration**
- **U.S.** cuts back (budget sequestration).  
- **Europe** flatlines amid austerity.  
- **China** keeps gradual, disciplined growth.

#### **2014–2021 — Crimea, China’s Rise, and Strategic Rivalry**
- **Russia:** sharp uptick after annexing Crimea (2014).  
- **NATO & Western Allies:** slow but steady rearmament.  
- **Asia:** Japan, India, and South Korea expand budgets amid regional flashpoints.  
- **China:** rises consistently, symbolizing its emergence as peer competitor.

#### **2022–2023 — Ukraine War & Indo-Pacific Tensions**
- **Russia:** explodes upward; war economy dominates GDP.  
- **Western Europe:** unprecedented defense increases (Germany’s Zeitenwende).  
- **U.S. and Asia:** strengthen deterrence.  
- **Global average spending** climbs — reversing 30 years of gradual disarmament.

---

**Overall Message:**  
The long arc (1993–2023) traces a complete global cycle —  
from **disarmament and peace dividend**, to **crisis-driven rearmament**, to **structural militarization**.  

- The **U.S. and its allies** lead major rearmament waves tied to geopolitical shocks.  
- **Russia** oscillates with regime strength and oil prices, culminating in wartime mobilization.  
- **China**’s smooth upward curve marks its transition from observer to great power.  
- **Middle Eastern powers** remain consistently high — conflict and energy rents anchor spending.  

This historical foundation sets the stage for forecasting:  
**How strongly does rival spending (especially from China or Russia) predict U.S. defense growth?**
"""
display(md(explanation))
```
